```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))

train_df = train_df.drop(['Unnamed: 0', 'index'], axis='columns')
train_df.head()

# train_df['loan_status'].value_counts()
```

```python
test_df = test_df.drop(['Unnamed: 0', 'index'], axis='columns')
test_df.head()
```

```python
# Convert categorical data to numeric and separate target feature for training data
dum_train = train_df.drop(['loan_status'], axis=1)
dum_train = pd.get_dummies(dum_train)
```

```python
# Convert categorical data to numeric and separate target feature for testing data
dum_test = test_df.drop(['loan_status'], axis=1)
dum_test = pd.get_dummies(dum_test)
```

```python
# find the missing variable/column by comparing the two dummy dfs
train_col = [x for x in dum_train.columns]

for col in dum_test:
    for x in train_col:
        if col == x:
            train_col.pop(train_col.index(x))

print(train_col)
```

```python
# add missing dummy variables to testing set

# the debt_settlment_flag_N values suggest that the debt_settlement_flag_Y values to be added should all be 0
dum_test['debt_settlement_flag_N'].value_counts()

# adding in the missing dummy variable column
dum_test['debt_settlement_flag_Y'] = 0
dum_test
```

```python
# training on logistic regression vs. random forest
```

Prediction: Random forest may do better because the data is currently unscaled, which may affect logistic regression's score relative to that of random forest's.

```python
# assign training data to x or y
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
X_train = dum_train
y_train = LabelEncoder().fit_transform(train_df['loan_status'])
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression()
log_model.fit(X_train, y_train)
```

```python
# define the testing x and y data
X_test = dum_test
y_test = LabelEncoder().fit_transform(test_df['loan_status'])
```

```python
print(log_model.score(X_train, y_train))
log_model.score(X_test, y_test)
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=1, n_estimators=500)
rf.fit(X_train, y_train)
```

```python
print(rf.score(X_train, y_train))
rf.score(X_test, y_test)
```

Random forest did indeed do better with both the training and test data sets. My hypothesis is that the uneven data scales in the dataframes has caused logistic regression to underperform

In the following steps, where data is scaled before re-fitted into the models, I think logistic regression will see a relative jump in score, enough to surpass the score from using random forest. Random forest should not be affected too much by scaled data.

```python
# Scale the data
from sklearn.preprocessing import StandardScaler as ss

scaler = ss().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_train_scaled

X_test_scaled = scaler.transform(X_test)
X_test_scaled
```

```python
# Train the Logistic Regression model on the scaled data and print the model score

log_model.fit(X_train_scaled, y_train)
print(log_model.score(X_train_scaled, y_train))
log_model.score(X_test_scaled, y_test)

```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

rf.fit(X_train_scaled, y_train)
print(rf.score(X_train_scaled, y_train))
rf.score(X_test_scaled, y_test)


```

My previous prediction turned out to be correct. The logistic regression approach became more effective -- even more effective than random forest -- once the data was scaled; random forest was not affected by much (though, in very exacting terms, its score when predicting test data dropped by ~0.001)
